# О MovieLens
MovieLens — это веб-система рекомендаций и виртуальное сообщество, которое рекомендует пользователям фильмы для просмотра на основе их предпочтений с помощью совместной фильтрации оценок и рецензий членов сообщества. Она содержит около 11 миллионов оценок примерно 8500 фильмов. MovieLens был создан в 1997 году GroupLens Research, исследовательской лабораторией факультета компьютерных наук и инженерии Университета Миннесоты, с целью сбора исследовательских данных и персонализированных рекомендаций.

## История
MovieLens не была первой системой рекомендаций, созданной GroupLens. В мае 1996 года GroupLens создала коммерческое предприятие под названием Net Perceptions, которое обслуживало таких клиентов, как E! Online и Amazon.com. E! Online использовала услуги Net Perceptions для создания системы рекомендаций для Moviefinder.com, а Amazon.com использовала технологию компании для формирования своей ранней системы рекомендаций для покупок потребителей.
Когда в 1997 году закрылся другой сайт с рекомендациями фильмов, eachmovie.org, исследователи, которые его создали, публично обнародовали 
анонимные данные собранных рейтингов, чтобы другие исследователи могли их использовать. Исследовательская группа GroupLens под руководством Брента Далена и Джона Херлокера использовала этот набор данных для запуска нового сайта рекомендаций фильмов, который они назвали MovieLens.

Затем применяет данную метрику к списку значений, сортирует и выводит словарь {x: value} по убыванию значения метрики value для данного x.  
Для вычисления метрик используются методы self._average, self._median, self._pvariance (population variance), вычисляющие метрики по известным формулам.

```python
%timeit -n 1 -r 1 ratings.movies.top_by_ratings(10)
ratings.movies.top_by_ratings(10)
```

N - количество фильмов, COLS - названия колонок данных. #### 1. Получение данных с сайта IMDB

Метод **get_imdb** позволяет получить информацию о фильме по movieId и списку полей. Метод проходит циклом for по списку movieId, загружая данные для соответствующего imdbId с сайта imdb.com при помощи метода self._parse_imdb или из кэша self.imdb_raw_data. Метод **_parse_imdb** отвечает за загрузку страницы для данного imdbID с помощью библиотеки requests и парсинг нужных полей с помощью BeautifulSoup. Загружаются поля с CSS-классами 'ipc-metadata-list__item' (элемент списка), 'ipc-metadata-list-item__label' (название поля) и 'ipc-metadata-list-item__list-content-item' (значения).
